```python
# Read CSV files
post = pd.read_csv('posts.csv')
following = pd.read_csv('following.csv')
notification = pd.read_csv('notifications.csv')
user = pd.read_csv('users.csv')
```

```python
# Drop unnecessary fields from the table
notification = notification.drop(columns = ['parent_comment_id','comment','sender_id','user_id','status','created_at','updated_at'])
```

```python
#Merge table post and notification together
df = pd.merge(post, notification, on='id')
```

```python
df.drop(columns = ['id'])
```

```python
# Select importtant features
features = ['user_id','title','content','post_id','action','type']
```

```python
def combine_features(row):
    str_user_id = str(row['user_id'])
    str_post_id = str(row['post_id'])
    return str_user_id +' '+ row['title']+' '+row['content']+' '+str_post_id+' '+row['action']+' '+row['type']
```

```python
for feature in features:
    df[feature] = df[feature].fillna('') #filling all NaNs with blank string
    df['combined_features'] = df.apply(combine_features,axis=1) #applying combined_features() method over each rows of dataframe and storing the combined string in ‚Äúcombined_features‚Äù column
```

```python
cv = CountVectorizer()
```

```python
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
cosine_sim = cosine_similarity(count_matrix)
```
